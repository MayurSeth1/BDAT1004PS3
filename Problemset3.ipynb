Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as mp

us_marriages = pd.read_csv('us-marriages-divorces-1867-2014.csv')

us_marriages.plot(x = 'Year', y = ['Marriages_per_1000', 'Divorces_per_1000'])

mp.xlabel("Year")

mp.ylabel("Marriage/Divorce Per Capita")

mp.legend(["Marriage Per Capita", "Divorce Per Capita"])

mp.show()
```

Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!


```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as mp

us_marriages = pd.read_csv('us-marriages-divorces-1867-2014.csv')

us_marriages = us_marriages.loc[us_marriages['Year'].isin([1900,1950,2000])]

marriage_per_capita = us_marriages['Marriages_per_1000']
marriage_per_capita = marriage_per_capita.values
divorce_per_capita = us_marriages['Divorces_per_1000']
divorce_per_capita = divorce_per_capita.values
year = us_marriages['Year']

df = pd.DataFrame({'Marriages_Per_Capita':marriage_per_capita, 'Divorces_Per_Capita':divorce_per_capita}, index=year)

df.plot.bar()

mp.xlabel("Year")

mp.ylabel("Marriages/Divorces Per Capita")

mp.legend(["Marriages Per Capita", "Divorces Per Capita"])

mp.show()
```

Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort 
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as mp

actor_kill_count = pd.read_csv("actor_kill_counts.csv")

actor = actor_kill_count['Actor']
Actor = actor.values

Count = actor_kill_count['Count']
Count = Count.values

df = pd.DataFrame({'Actor' : Actor, "Count" : Count})
df = df.sort_values('Count')

mp.barh('Actor', 'Count',data=df)

mp.xlabel("Kill Count")

mp.show()
```
